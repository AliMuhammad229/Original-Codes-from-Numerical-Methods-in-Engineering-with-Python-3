```python
# Functions and Modules
from math import atan
def finite_diff(f,x,h=0.0001):
    df =(f(x+h)-f(x-h))/(2.0*h)
    ddf = (f(x+h)-2.0*f(x)+f(x-h))/h**2
    return df,ddf
x = 0.5
df,ddf = finite_diff(atan,x)
print('First derivative = ',df)
print('Second derivative = ',ddf)
```
